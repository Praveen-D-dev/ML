```python
import numpy as n
import seaborn as s
import matplotlib.pyplot as m
```

```python
t=s.load_dataset("tips")
```

```python
s.scatterplot(x="day",y="total_bill",data=t)
m.title("Scatterplot of Total Bill vs Tip")
m.xlabel("day")
m.ylabel("total_bill")
m.show()
```

```python
s.lineplot(x="day",y="total_bill",data=t)
m.title("LinePlot of Total Bill vs Tip")
m.xlabel("day")
m.ylabel("total_bill")
m.show()
```

```python
s.barplot(x="day",y="total_bill",data=t)
m.title("Barplot of Total Bill vs Tip")
m.xlabel("day")
m.ylabel("total_bill")
m.show()
```

```python
s.boxplot(x="day",y="total_bill",data=t)
m.title("boxplot of Total Bill vs Tip")
m.xlabel("day")
m.ylabel("total_bill")
m.show()
```

```python
s.violinplot(x="day",y="total_bill",data=t)
m.title("Violinplot of Total Bill vs Tip")
m.xlabel("day")
m.ylabel("total_bill")
m.show()
```

```python
s.countplot(x="day", data=t)
m.title("Count of Tips by Day")
m.xlabel("Day")
m.ylabel("Count")
m.show()
```

```python
s.histplot(data=t, x="total_bill", kde=True)
m.title("Distribution of Total Bill")
m.xlabel("Total Bill")
m.ylabel("Frequency")
m.show()
```

```python
s.jointplot(data=t, x="total_bill", y="tip", kind="reg")
m.suptitle("Joint Plot of Total Bill and Tip with Regression Line", y=1.02)
m.show()
```

```python
s.pairplot(t, hue="sex")
m.suptitle("Pair Plot of Numerical Variables by Sex", y=1.02)
m.show()
```

```python
s.pointplot(x="day", y="total_bill", hue="smoker", data=t)
m.title("Point Plot of Total Bill by Day and Smoker Status")
m.xlabel("Day")
m.ylabel("Total Bill")
m.show()
```

```python
s.FacetGrid(t, col="time").map(s.histplot, "total_bill")
m.suptitle("Distribution of Total Bill by Time of Day", y=1.02)
m.show()
```

```python
n=t
co=n.select_dtypes(include='number').corr()
s.heatmap(co,annot=True,cmap='coolwarm')
m.title("Correlation Heatmap")
m.show()
```
